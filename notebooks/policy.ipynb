```python
with Apps.init() as apps:
    apps.where(
        lambda app: app.org == "ttd"
        and app.policy[
            '//xacml:ObligationExpression/xacml:AttributeAssignmentExpression[@Category="urn:altinn:minimum-authenticationlevel"]/xacml:AttributeValue/text()',
            :,
        ]
        .map(lambda value: int(value.text) if value.text is not None else 0)
        .some(lambda level: level == 0)
    ).select({"Repo": lambda app: app.repo_url}).text_table()
```

```python
with Apps.init() as apps:
    apps.where(lambda app: app.env == "prod").select(
        {
            "Policy": lambda app: app.policy[
                "//xacml:AnyOf[.//xacml:AttributeValue[text()='write'] and (.//xacml:AttributeValue[text()='sign'] or .//xacml:AttributeValue[text()='confirm'])]",
                :,
            ]
        }
    ).where(lambda app: app["Policy"].length > 0).table()
```
