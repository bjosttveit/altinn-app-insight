## Missing icu-data-full when using DisplayFooter

```python
# RUN apk add --no-cache icu-libs icu-data-full tzdata
with Apps.init() as apps:
    apps.where(
        lambda app: app.dockerfile.find(r"RUN apk add.*icu-libs") != None
        and app.dockerfile.find(r"RUN apk add.*icu-data-full") == None
        and app.app_settings.some(lambda app_settings: app_settings[".PdfGeneratorSettings.DisplayFooter"] == True)
    ).select(
        {
            "Dockerfile": lambda app: app.dockerfile.find(r"RUN apk add.+icu-libs"),
            "PdfGeneratorSettings": lambda app: str(
                app.app_settings.filter(
                    lambda app_settings: app_settings[".PdfGeneratorSettings.DisplayFooter"] == True
                )
                .map(lambda app_settings: app_settings[".PdfGeneratorSettings"])
                .first
            ),
        }
    ).order_by(
        lambda app: (app.org, app.app, app.env)
    ).table()
```
